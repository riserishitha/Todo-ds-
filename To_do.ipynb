```python
tasks = []

while True:
    print("\n--- TO-DO LIST APP ---")
    print("1. Add Task")
    print("2. Remove Task")
    print("3. View Tasks")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        task = input("Enter task to add: ")
        tasks.append(task)
        print("Task added!")

    elif choice == "2":
        task = input("Enter task to remove: ")
        if task in tasks:
            tasks.remove(task)
            print("Task removed!")
        else:
            print("Task not found.")

    elif choice == "3":
        print("\nYour Tasks:")
        for t in tasks:
            print("- " + t)

    elif choice == "4":
        print("Exiting program...")
        break

    else:
        print("Invalid choice, try again!")
```
